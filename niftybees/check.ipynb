```python
import pandas as pd
```

```python
data = pd.read_csv('AUROPHARMAEQN.csv',usecols=[0,2,9])
```

```python
def buy():
    global amount,p,flags,rate,depth
    try:
        money = amount/(depth-len(flags))
        amount -= money
        quantity=money//p
        amount += money%p
        flags.append({'cost':p,'qty':quantity})
        return p*(1+rate),p*(1-rate)
    except:
        return target,sl

def sell():
    global flags,amount,p,rate,depth
    a=flags.pop()
    amount+=p*a['qty']
    q=0
    try:
        q=flags[-1]['cost'] 
    except: pass
    return q*(1+rate),q*(1-rate)
```

```python
amount = 500000
rate=.02
depth=5
records=[]
flags = []
for p in data.Average_Price:
    action=None
    if len(flags) == 0:
        target,sl=buy()
        action='buy'
    elif p>=target:
        target,sl=sell()
        action='sell'
    elif p <= sl :
        target,sl = buy()
        action='buy'
    records.append([amount,action,amount+sum([i['qty'] for i in flags ])*p,len(flags),str(flags)])
    
```

```python
final=data.join(pd.DataFrame(records,columns=['amount','action','currentvalue','depth','flags']))
```
